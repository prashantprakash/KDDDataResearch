```python
import matplotlib.pyplot as plt; plt.rcdefaults()
import numpy as np
import matplotlib.pyplot as plt
 
objects = ('2', '3', '4', '5', '6', '10','20','30')
y_pos = np.arange(len(objects))
accuracy = (0.9296685, 0.8888135, 0.80711967, 0.77198124, 0.930633, 0.9287522,0.82988614,0.93157357)
 
plt.bar(y_pos, accuracy, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Accuracy')
plt.xlabel('K of second level clusters')
plt.title('K of second_level_clusters vs Accuracy')
 
plt.show()
```
